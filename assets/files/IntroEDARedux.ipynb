How do we read in our film scripts dataset? How would we sort the values in the dataset? How could we remove all the null values from the dataset? How we group our data by year?

```python
import requests
import pandas as pd
```

```python
film_scripts = pd.read_csv('pudding_film_scripts.csv')
```

```python
film_scripts.head()
```

```python
film_scripts.sort_values(by=['year'], ascending=False, inplace=True)
```

```python
film_scripts = film_scripts.rename(columns={'gross (inflation-adjusted)' : 'gross_ia'})
```

```python
film_scripts.isna().any()
```

```python
subset_film_scripts = film_scripts.dropna()
```

```python
film_years = film_scripts.groupby('year')
```

```python
film_years.get_group(1980)
```

```python
film_scripts.fillna(0, inplace=True)
```

```python
grouped_films = film_scripts.groupby('year')['title'].count().reset_index()
```

```python
grouped_films = grouped_films.rename(columns={'title' : 'number_of_films'})
grouped_films.head()
```

```python
grouped_films.columns, film_scripts.columns
```

```python
result = pd.merge(grouped_films, film_scripts, on="year", how='outer')
```

```python
result
```

```python
result.plot(x='number_of_films', y='gross_ia', kind='scatter')
```
